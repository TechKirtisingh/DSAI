### NUMPY :
NumPy is a Python library used for working with arrays.

It also has functions for working in domain of linear algebra, fourier transform, and matrices. 

In Python we have lists that serve the purpose of arrays, but they are slow to process.

NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

```python
import numpy as np

list = [1,2,3,4,5]
arr = np.array(list)
print("Array : " , arr)
print("\n")


arr2 = np.array([[1,2,3] , [4,5,6]])
print("2D arrya: \n\n" , arr2)

arr3 = np.array([[[1,2,3] , [4,5,6]] , [[11,12,13] , [14,15,16]]])
print("3D array: \n\n" , arr3)

print("\n" ,arr2[1,1])
print("\n" , arr3[1 ,0 ,0])

print("Dimention of the array: " , np.ndim(arr3)) # gives the dimension of the array

print("Size gives the number of elements in the array:\n ")

print(arr3.size)
print(arr2.size)
print(arr.size)


print("Shapes of the arrays: \n")

print(arr3.shape)
print(arr2.shape)
print(arr.shape)


print("Slicing of the array: \n")
print(arr[1:3])

print(arr[::-1]) # reverse the array

is_palindrome = np.array_equal(arr , arr[::-1]) 
print("Is the array a palindrome? : " , is_palindrome)


```

```python
a = np.array([2,4,6,8] , dtype = float)


print(a)
print(a+2)
print(a*2)
print(a**2)

```

```python
myarr = np.zeros([2,3])
print ("Array of zeros: \n\n" , myarr)

myarr1 = np.ones([2,3])
print ("Array of ones: \n\n" , myarr1)

myarr3 = np.full([2,3] , 5) # full array with 5
print ("Array of full: \n\n" , myarr3)

myarr2 = np.arange(1,10,2) # start , stop , step
print("Array of range: \n\n" , myarr2)


print("\n" , np.linspace(1,10,5)) # start , stop , number of elements // linspace is used to create an array of evenly spaced numbers over a specified interval.
```

```python
# average of the array

arr = np.array([1,2,3,4,5])
print("Average of the array: " , np.mean(arr))
```

```python
# random numbers

ar = np.random.rand(3,2) # 3 rows and 2 columns
print("\nRandom numbers: \n\n" , ar)
print("\nMaximum value in the array: \n\n" , np.max(ar))
print("\nMinimum value in the array: \n\n" , np.min(ar))
print("\nSum of the array: \n\n" , np.sum(ar))
print("\nAverage of the array:\n\n " , np.mean(ar))

ar2 =np.array([5,6,9,8,4])
print("\n" ,ar2.max())
print("\n",ar2.min())
print("\n" ,ar2.sum())
print("\n" ,ar2.mean())

print("\n Identitiy Matrix\n : ")

ar3 = np.eye(3) # identity matrix
print(ar3)
```

```python
# absolute value : give positve value
array =  [2,5,7,8,9]

print(np.abs(array)) # absolute value
print("\n" , np.sqrt(array)) # square root
print("\n" , np.square(array)) # square
print("\n" , np.log(array)) # logarithm
```
