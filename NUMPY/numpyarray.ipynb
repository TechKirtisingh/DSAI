```python
import numpy as np

# Matrix A
A = np.array([
    [2, 4, 6],
    [8, 10, 12],
    [14, 16, 18]
])

# Matrix B
B = np.array([
    [1, 3, 5],
    [7, 9, 11],
    [13, 15, 17]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
print("Shapes of A and B:", A.shape, B.shape)

print("\n Size of A and B:", A.size, B.size)

print(f"\n Dimensions of A : {A.ndim} , and B : {B.ndim}")

print("\n Addition :" , A+B)

print("\n Substraction :" , A-B)

print("\n Element-wise Multiplication :" , A*B)

print("\n Multiplication :" , np.matmul(A,B))

print(f"Transpose of matrix A: {A.T} , and B : {B.T}")

print(f"Determinant of matrix A: {np.linalg.det(A)} , and B : {np.linalg.det(B)}")


```

```python
# not possible

# print("Inverse of Matrix A:\n", np.linalg.inv(A))
```

```python
print("Trace of matrix A:", np.trace(A))

# print(A.trace())

print("Trace of matrix B:", np.trace(B))
# print(B.trace())
```

```python
print("Maximum value in matrix A:", np.max(A))
print("Minimum value in matrix A:", np.min(A))
  
print(f"Sum of matrix A: {A.sum()}, Sum of matrix B: {B.sum()}")

print("Mean of matrix A:", np.mean(A))
```

```python
print("Row wise sum of matrix B:", np.sum(B, axis=1))
print("Column wise sum of matrix B:", np.sum(B, axis=0))
```

```python
print("Extract First row of matrix A:", A[0, :])

print("Second column of matrix B:", B[:, 1])

print("Extact last row " , A[-1])

print("Bottom right element :" , A[-1 , -1])


```

```python
print("Flattened matrix B:", B.flatten())  # flatten matrix means convert 2D array to 1D array

print("Reshape A to 1D array:", A.reshape(1,9))
```

```python
# find Enginevalue and Eigenvector of a matrix A

print("Eigenvalues and Eigenvectors of matrix A:")

print(np.linalg.eig(A))  # returns a tuple of eigenvalues and eigenvectors
```

```python
# 15. Concatenate A and B:
    #  a) Horizontally (hstack)
    #  b) Vertically (vstack)
    
    
print("Horizontally concatenated A and B:\n", np.hstack((A, B)))
print("\nVertically concatenated A and B:\n", np.vstack((A, B)))
```
