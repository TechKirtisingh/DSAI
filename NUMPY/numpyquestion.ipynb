```python
import numpy as np
```

```python
emp_salary = np.zeros(10)
print( emp_salary)

for i in range(10):
     emp_salary[i]= int(input(f"Enter the employee salary: {i+1} "))

emp_salary = emp_salary*1000
print("\nEmployee salaries are: " ,  emp_salary)

print("\nTotal salary Expenses: " , np.sum(emp_salary))

print("\nAverage salary Expenses: " , np.mean(emp_salary))

print("\nHighest Salary Expenses: " , np.max(emp_salary))

print("\nLowest Salary Expenses: " , np.min(emp_salary))

print("\nCount of employees with salary greater than 50000: " , np.sum(emp_salary > 50000))

print("\n Count of employees with salary less than 30000:" , np.sum(emp_salary < 30000))

specific_salary = int(input("\nEnter the specific salary to check: ")) * 1000
print(f"\nCount of employees with salary equal to {specific_salary}: ", np.sum(emp_salary == specific_salary))

print("\nSort employee salaries in ascending order: " , np.sort(emp_salary))

print("\nSort employee salaries in descending order: " , np.sort(emp_salary)[::-1])

print("\n Increse Salary by 10% : " ,  emp_salary * 1.1)

replace_salary = int(input("\nEnter the salary to be replaced: ")) * 1000
new_salary = int(input("\nEnter the new salary: ")) * 1000
emp_salary[emp_salary == replace_salary] = new_salary


print("\nUpdated employee salaries are: " ,  emp_salary)


```
