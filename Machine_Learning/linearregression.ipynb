## Linear Regression

(Gives Numeric Solutions)

When We Want a numeric Value according to a single variable like price (Some prices are also given) -----------> Rooms are given

Predict Price if any number of rooms are given 

This is called Single Linear Regression .

If there are Multiple Variables Given

Price --------------------> Rooms, area, Floor etc.
Now Prediction becomes Tougher.

This is calles Multiple Linear Regression .


You 1st import sklearn.linear_model as LR
Take the Dataframe
X= Independent Variable (No. of Rooms)
Y = Dependent Variable (Price That you want to predict)
model = LR()
model.fit(X,Y)  This function tries to find the Best Fit line in the Graph for predictions

### AREA VS PRICE

```python
import pandas as pd
import sklearn.linear_model as lm

data = {
    "Area" : [1000, 2000, 3000, 4000, 5000],
    "Price" : [20,40,60,80,100]
}

df = pd.DataFrame(data)
X= df[["Area"]]
Y= df["Price"]
model = lm.LinearRegression()
model.fit(X,Y)
predictPrice = model.predict([[2210]])
print("Price Of House : ", predictPrice)
```

### EXPERIENCE VS SALARY

```python
import pandas as pd
import sklearn.linear_model as lm

df = pd.DataFrame({})
df = pd.read_csv("experience_salary_40_records.csv")
# print(df)

X= df[['Experience']]
Y= df[['Salary']]

model = lm.LinearRegression()

model.fit(X,Y)

exp= int(input("Years Of Experience : "))
predict_Salary = model.predict([[exp]])

print(f"Salary for {exp} years : {predict_Salary}")
```
