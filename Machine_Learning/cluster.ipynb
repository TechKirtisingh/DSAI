```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create DataFrame
df = pd.DataFrame({
    
    "Age" : [20 ,22,25,27,30,32,35,38,40,42],
    "Income": [25000  , 27000 , 29000 , 31000 , 60000 , 62000 , 64000 , 66000 ,68000 , 70000]
    
})

print(df)

print("--"*100)

# Select feature

X = df[["Age" , "Income"]]

# Create Model
model = KMeans(n_clusters = 2 , random_state=42)

# Train Model
df["Cluster"] = model.fit_predict(X)

#Display Data Frame
print(df)

##### PLOT CLUSTERS

plt.scatter(df["Age"], df["Income"], c=df["Cluster"], cmap='viridis')  # cmap='viridis' is used to color the clusters 
plt.title("KMeans Clustering of Age and Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```
