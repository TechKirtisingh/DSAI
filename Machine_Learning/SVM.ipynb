## What is Fitting?

Fitting means how well a machine learning model learns from the training data.

# overfitting

Overfitting tab hota hai jab model data ko samajhne ke bajay yaad (memorize) kar leta hai.

Training mein bahut achha perform karta hai, lekin naye data par fail ho jata hai.

# underfitting

Underfitting tab hota hai jab model bahut simple hota (data hi km provide kiye hai) hai.

Wo data ka pattern hi nahi samajh pata.

Isliye training aur testing dono mein poor performance deta hai.
