### ML learns in five simple step:

1. Collect Data

2. Observe Patterns

3. Build a Model : After observing patterns, the algorithm creates a mathematical model. 

4. Test on new data

5. Improve :  If the prediction is wrong, the algorithm adjusts itself



##### WHY DOES ML NEED LOTS OF DATA? 

> More good-quality data → Better learning.  
> Poor or very little data → Poor predictions.

####  SUPERVISED LEARNING 

In Supervised Learning, the computer learns from labeled data. 

Labeled data means every input already has the correct answer (output). 

Formula ::  Input + Correct Output → Learn → Predict New Output


####   UNSUPERVISED LEARNING 

In Unsupervised Learning, the data has no labels. 

The computer must discover hidden patterns or groups by itself. 

Formula :

Input Only 
↓ 
Find Similarities 
↓ 
Create Groups (Clusters) 


####  SEMI-SUPERVISED LEARNING

In Semi-Supervised Learning, only some data is labeled, while most data is unlabeled. 
Formula::
 
Small Labeled Data + Large Unlabeled Data 
 ↓
Learning 


##### Labeling data can be expensive and time-consuming, so semi-supervised learning makes use of the large amount of unlabeled data that is often available. 

#### REINFORCEMENT LEARNING 

In Reinforcement Learning, an agent learns by interacting with an environment. 

It receives: •    

Reward for a good action.  

•Penalty for a bad action. 

The goal is to maximize rewards over time. 



#### Type 

Supervised Learning :: A teacher gives questions and answers

Unsupervised Learning :: Students group themselves without any teacher. 

Semi-Supervised Learning :: A few students have answer keys, and the rest learn with their help. 

Reinforcement Learning :: Learning by trial and error with rewards and penalties, like playing a game.
