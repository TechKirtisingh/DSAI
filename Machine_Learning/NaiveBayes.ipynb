#### NAIVE BAYES ALGORITHM::

It predicts the class of a new data point using probability and Bayes' Theorem. 

Naive Bayes calculates:

 "Which class has the highest probability?" 
 
 The class with the highest probability becomes the prediction. 

### CLUSTERING

Clustering is an Unsupervised Learning technique that groups similar data points together.

Isme data ke labels (answers) pehle se nahi hote.

Algorithm khud hi similar data ko ek group (cluster) mein daal deta hai.
