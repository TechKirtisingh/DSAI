### Q15. Create a stack plot showing cumulative marks of students grouped by three Courses.

```python
courses = df["Course"].unique()[:3]

course1 = df[df["Course"] == courses[0]]["Marks"].reset_index(drop=True)
course2 = df[df["Course"] == courses[1]]["Marks"].reset_index(drop=True)
course3 = df[df["Course"] == courses[2]]["Marks"].reset_index(drop=True)

n = min(len(course1), len(course2), len(course3))
x = range(n)

plt.figure(figsize=(10,5))
plt.stackplot(
    x,
    course1[:n],
    course2[:n],
    course3[:n],
    labels=courses
)

plt.legend(loc="upper left")
plt.title("Stack Plot of Marks by Course")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

### Q16. Create a line chart showing the average Marks for each Age, sorted by Age.

### Q17. Plot two lines on the same graph
Average Marks by Course
Maximum Marks by Course

### 

```python
result = df.groupby("Course")["Marks"].agg(["mean", "max"])

result.plot(
    kind="line",
    marker="o"
)

plt.title("Average and Maximum Marks by Course")
plt.xlabel("Course")
plt.ylabel("Marks")
plt.legend(["Average", "Maximum"])
plt.show()
```

### Q18. Create a grouped bar chart

### Q19. Top 10 students by Marks

```python
top10 = df.sort_values("Marks", ascending=False).head(10)

colors = ["green" if m >= 85 else "red" for m in top10["Marks"]]

plt.bar(top10["Student_ID"], top10["Marks"], color=colors)

plt.title("Top 10 Students by Marks")
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.show()
```
