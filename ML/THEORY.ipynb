 What is Descriptive Statistics? 
 Descriptive Statistics is a branch of statistics that is used to collect, organize, summarize, analyze, and present data in a meaningful way.

 hoe it is helpful in ml (statics): 

Using descriptive statistics, we can easily answer questions like:
• What is the average customer age? 
• What is the highest salary?
• What is the lowest mark?  
• How much variation exists in sales?
• Which product is sold the most?

#### HYPOTHESIS
hyphothesis is a assumption ya statement hota hai jise hum data ki help se test karte hain ki ye sahi hai ya galat.

Hypothesis Testing is a method of checking whether a claim is true or false using sample data. 

### Types of Hypothesis
1. Null Hypothesis (H₀)

Null means Nothing has changed. or The claim is true.

2. Alternative Hypothesis (H₁) 
Alternative means Something is different.

Correlation batata hai ki do variables (X aur Y) ke beech relationship kaisa hai.

#### Correlation Coefficient (r)
