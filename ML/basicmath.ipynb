```python
import numpy as np

marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])

print("Dataset:")
print(marks)
```

```python
# mean
print("\n Mean : " , np.mean(marks))

# Median
print("\nMedian : " , np.median(marks))
```

```python
# mode

from scipy import stats
print("\n Mode :", stats.mode(marks).mode)



# max and min valuees
print(f"\n Maximum value : {np.max(marks)} , and Minimum value : {np.min(marks)}")
```

```python
# range of data set
# Range Formula Range = Maximum − Minimum 

print(f"\n Range of dataset : {np.max(marks) - np.min(marks)}" )

```

```python
# varience of data set

print(f"\n Variance of dataset : {np.var(marks)}" )


# standard deviation of data set

print(f"\n Standard Deviation of dataset : {np.std(marks)}" )


# Calculate the 25th percentile (Q1).
q1 = np.percentile(marks, 25)
print(f"\n 25th Percentile (Q1): {q1}")

# 9. Calculate the 50th percentile (Q2/Median).

q2 = np.percentile(marks, 50)
print(f"\n 50th Percentile : {q2}")

# Calculate the 75th percentile (Q3).
q3 = np.percentile(marks, 75)
print(f"\n 75th Percentile (Q3): {q3}")
```
