```python
import pandas as pd
```

```python
list = ['rk' , 'vk' , 'ak' , 'bk']
s=pd.Series(list) #  Create a Series
s=pd.Series(list , index=['a', 'b', 'c', 'd']) #  Create a Series with custom index
print(s) 
```

```python
d = {
    "RollNo" : [1,2,3,4,5 ,6,7],
    "name" : ['rk' , 'vk' ,'sk', 'ak' , 'bk' , 'ck' ,'dk'] ,
    "marks" : [56 , 80 , 70 , 60 ,50 , 90 , 100] ,
    "branch" : ["cs" , "cs" , "DS" ,"DS" , "cs" , "cs" , "DS"]
}
print(d)
print(type(d))
df = pd.DataFrame(d) #  Create a DataFrame
print(df)
```
