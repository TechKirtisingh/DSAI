```python
# graph between sales and products
# x_label : means the label of x axis
# x means the data of x axis
# y_label : means the label of y axis
# y means the data of y axis

import matplotlib.pyplot as plt

sales_df.plot(kind = "bar", title = "Sales of Products" , color = "red" , figsize = (10,5) , xlabel = "Products" , ylabel = "Sales" , x = "Product" , y = "Sales" )

sales_df.plot(kind = "pie", title = "Sales of Products" , color = "blue" , figsize = (10,5) , xlabel = "Products" , ylabel = "Sales" , x = "Product" , y = "Sales" )
```

```python
# in this graph we are going to show the total sales of products. so we have to group the data by product and then sum the sales of each product and then plot the graph.

sales_df.groupby("Product")["Sales"].sum().plot(
    kind="bar",
    title="Total Sales of Products",
    color="red",
    figsize=(10,5),
    xlabel="Products",
    ylabel="Total Sales"
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame({
    "Sales": [2020, 3090, 4040, 5550, 6000, 7070, 8800],
    "Month": ["June", "July", "August", "September", "October", "November", "December"]
})

# alpha = 0.3 means the transparency of the line.  (lie betwwen 0 and 1) 0 means completely transparent and 1 means completely opaque.
df.plot(kind = "line" , x = "Month" , y = "Sales" , title = "Sales of Products" , color = "purple" , xlabel = "Month" , ylabel = "Sales" , grid = True , alpha = .3 , markeredgewidth = 2 , marker = 'o' , linestyle = "--")
plt.gca().set_facecolor('skyblue')
plt.Figure(figsize=(10, 8) , facecolor = "lightgreen")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame({
    "Sales": [2020, 3090, 4040, 5550, 6000, 7070, 8800],
    "Month": ["June", "July", "August", "September", "October", "November", "December"]
})


# Create plot
ax = df.plot(
    x="Month",
    y="Sales",
    kind="line",
    figsize=(10, 5),
    color="red",
    title="Monthly Sales",
    marker="o",
    markeredgewidth=4 
    
)

# Customize graph
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.grid(True)

plt.gca().set_facecolor('skyblue')  # Set background color to light gray

plt.show()
```
