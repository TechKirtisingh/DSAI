```python
# graph between sales and products
# x_label : means the label of x axis
# x means the data of x axis
# y_label : means the label of y axis
# y means the data of y axis

import matplotlib.pyplot as plt

sales_df.plot(kind = "bar", title = "Sales of Products" , color = "red" , figsize = (10,5) , xlabel = "Products" , ylabel = "Sales" , x = "Product" , y = "Sales" )

sales_df.plot(kind = "pie", title = "Sales of Products" , color = "blue" , figsize = (10,5) , xlabel = "Products" , ylabel = "Sales" , x = "Product" , y = "Sales" )
```

```python
# in this graph we are going to show the total sales of products. so we have to group the data by product and then sum the sales of each product and then plot the graph.

sales_df.groupby("Product")["Sales"].sum().plot(
    kind="bar",
    title="Total Sales of Products",
    color="red",
    figsize=(10,5),
    xlabel="Products",
    ylabel="Total Sales"
)
```
