### Create a class Student with attributes Roll Number, Name, and Marks. Create an object and display the student's details


```python
class Student:
    def __init__(self , rollno , name , marks):
        self.rollno = rollno
        self.name = name
        self.marks = marks
        
    def display_details(self):
        print(f"Roll Number: {self.rollno}")
        print(f"Name: {self.name}")
        print(f"Marks: {self.marks}")
        
s1 = Student(101 , "Ram", 85)
s1.display_details()

s2 = Student(102 , "Shyam", 90)
s2.display_details()
```

### Create a class Employee with attributes Employee ID, Name, and Salary. Create an object and display the employee information.

### Create a class BankAccount with attributes Account Number and Balance. Implement methods to deposit money and display the updated balance. Create an object and test the methods

```python
class BankAccount:
    def __init__(self, Accountno, Balance):
        self.Accountno = Accountno
        self.Balance = Balance
    def deposit(self, amount):
        self.Balance += amount
        print("\n")
        print(f"Deposited: {amount}             Updated Balance: {self.Balance}") 
        print("\n")

    def display(self):
        print(f"Account Number: {self.Accountno}            Balance: {self.Balance}") 

acc1 = BankAccount(1001, 5000) 
acc1.display()
acc1.deposit(1000)
acc1.display()
```

```python
# Write a Python program to find the sum of the first N natural numbers.
n=int(input("Enter a number: "))
sum=0
for i in range (1, n+1):
    sum = sum + i
    
print(f"the sum of first {n} natural numbers is: {sum}")
```

```python
# Write a Python program to print the multiplication table of a given number.

n= int(input("Enter a number of which you want to print the table"))

for i in range (1,11):
    print(f"{n} x {i} = {n*i} \n")
    
```

```python
# Write a Python program to calculate the factorial of a number using a loop.

n = int(input("Enter a number to calculate its factorial: "))

fact = 1

for i in range(1,n+1):
    fact = fact * i
    
print(f"The factorial of {n} is: {fact}")
    
```

```python
# Write a Python program to print the Fibonacci series up to N terms.

n = int(input("Enter the number of terms for Fibonacci series: "))

firstnum , secondnum = 0,1

print("Fibonacci Series: \n")
for i in range(n):
    
    print(firstnum, end=' ')

    firstnum, secondnum = secondnum, firstnum + secondnum
    
```

## Conditional Statement Programs
Write a Python program to check whether a number is even or odd.

Write a Python program to determine whether a student has passed or failed based on marks.

```python
# Write a python Program to check wether a number is even or odd.

num = int(input("Enter a number to check if it is even or odd: "))

def even_odd(num):
    
    if num % 2 == 0:
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")
        
even_odd(num)
```

```python
# Write a Python program to find the largest among three numbers.

n1 , n2 , n3 = map(int , input("Enter three numbers separated by space: ").split())

if n1 >= n2 and n1 >= n3:
    largest = n1    
elif n2 >= n1 and n2 >= n3:
    largest = n2
else:
    largest = n3
    
print(f"The largest number among {n1}, {n2}, and {n3} is: {largest}")
```

```python
# write a python program to check weather a year is a leap year or not.

year = int(input("Enter a year to check if it is a leap year or not: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")   
    
else:
    print(f"{year} is not a leap year.") 
    
```

```python
# Write a Python program to determine whether a student has passed or failed based on marks.

marks = int(input("Enter the marks obtained by the student: "))

def check_pass_fail(marks):
    if marks >= 40:
        print("The student has passed.")
    else:
        print("The student has failed.")
        
check_pass_fail(marks)
```

```python
# Write a Python program to calculate the grade of a student based on percentage.

percentage = float(input("Enter the percentage obtained by the student: "))

def clac_grade(percentage):
    if percentage >= 90:
        grade = 'grade A'
    elif percentage >= 80:
        grade = 'grade B'
    elif percentage >= 70:
        grade = 'grade C'
    elif percentage >= 60:
        grade = 'grade D'
    else:
        grade = 'grade F'
        
    print(f"The grade of the student based on {percentage}% is: {grade}")

clac_grade(percentage)
```
