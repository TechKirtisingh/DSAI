```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120],

    "Name": ["Aman","Riya",None,"Karan","Priya","Rahul","Neha",None,
             "Vikas","Pooja","Ankit","Simran","Rohit",None,"Kajal",
             "Deepak","Meena","Arjun","Nisha",None],

    "Age": [20,21,np.nan,22,19,23,21,20,np.nan,22,
            24,23,21,20,22,np.nan,19,21,23,20],

    "City": ["Delhi","Mumbai","Pune",None,"Jaipur","Delhi","Mumbai",
             "Pune","Delhi",None,"Jaipur","Mumbai","Delhi","Pune",
             None,"Delhi","Mumbai","Jaipur","Pune","Delhi"],

    "Marks": [85,90,78,np.nan,88,76,91,82,95,np.nan,
              80,84,79,92,87,75,np.nan,89,81,93]
})

print(df1)
```

```python
df1.fillna(value={'Name': 'Unknown', 'Age': 0, 'City': 'Not Avialable', 'Marks': 0}, inplace=True)  
print(df1)
```

```python
mean_marks = df1['Marks'].mean()# to find the mean of marks column
print("Mean of Marks column:", mean_marks)
```

```python
# replace....

df1['Marks'].replace(to_replace= 0 , value = mean_marks, inplace=True)  # to replace the 0 value with mean of marks column
print(df1)
```
