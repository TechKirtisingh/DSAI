### plotly

an open-source, browser-based graphing library used to create interactive, publication-quality data visualizations.

Plotly is a Python library used to create interactive graphs.

Core Modules::

plotly.express (PX): The high-level wrapper. It creates entire, pre-styled charts in a single line of code using Pandas DataFrames.

plotly.graph_objects (GO): The low-level API. It gives you absolute control over individual elements, layouts, and data structures.

```python
from IPython.display import HTML, display
```

```python
import plotly.express as px

import pandas as pd

df = pd.DataFrame({
    "Order_ID": range(101, 121),
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop",
        "Mobile","Tablet","Laptop","Mobile","Tablet",
        "Laptop","Mobile","Tablet","Laptop","Mobile"
    ],
    "Category": [
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "City": [
        "Delhi","Mumbai","Lucknow","Delhi","Lucknow",
        "Mumbai","Delhi","Lucknow","Mumbai","Delhi",
        "Lucknow","Mumbai","Delhi","Lucknow","Mumbai",
        "Delhi","Lucknow","Mumbai","Delhi","Lucknow"
    ],
    "Sales": [
        65000,30000,25000,70000,32000,
        27000,68000,35000,26000,72000,
        33000,28000,75000,34000,29000,
        76000,36000,30000,78000,37000
    ],
    "Profit": [
        12000,5000,4500,13000,5500,
        4700,12500,6000,4800,14000,
        6200,5000,14500,6300,5200,
        15000,6500,5400,15500,6800
    ],
    "Quantity": [
        5,10,8,6,9,
        7,5,11,8,6,
        10,7,5,9,8,
        4,12,9,5,10
    ]
})

print(df)

```

```python
fig = px.scatter(df,
                 x="Sales",
                 y="Profit",
                 title="Sales vs Profit",
                 color="Category")

# fig.show(renderer="browser")
display(fig)
```

```python
total_sale = df.groupby("Product")["Sales"].sum().reset_index()

fig = px.bar(total_sale, x="Product", y="Sales",color = 'Product' , title="Total Sales by Product")
# fig.show(renderer="browser")
display(fig)
```

```python
fig = px.pie(total_sale,
             names='Product',
             values='Sales',
             title='Sales Distribution by Product',
             color='Product'
)

# fig.show(renderer="browser")

display(fig)
```

```python
fig = px.scatter(
    df,
    x='Sales',
    y='Profit',
    size='Quantity',
    color='Product',
    title='Sales vs Profit'
)

# fig.show(renderer="browser")

display(fig)
```

```python
fig = px.histogram(df, x="Sales", nbins=10, title="Sales Distribution")
# fig.show(renderer="browser")
display(fig)
```

```python
# sales distribution of each product using box plot 
from IPython.display import HTML, display

fig = px.box(df,
             x='Product',
             y='Sales',
             title='Sales Distribution by Product'
)

display(fig)
# fig.show(renderer="browser")
```

```python
#6. Using groupby('City'), calculate the total Sales of each city and create a bar chart.

total_sale_city = df.groupby("City")["Sales"].sum().reset_index()   

fig = px.bar(total_sale_city, x="City", y="Sales", color='City', title="Total Sales by City")

display(fig)

```

```python
avg_prof = df.groupby('Product')['Profit'].mean().reset_index()

fig = px.line(avg_prof, x='Product', y='Profit', title='Average Profit by Product')

display(fig)
```
