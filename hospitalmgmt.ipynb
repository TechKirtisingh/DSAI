Develop a Hospital Management System using Multilevel Inheritance.

Requirements:

Create a class Hospital.

Constructor should initialize:

Hospital Name
City

Create a class Doctor inheriting from Hospital.

Constructor should initialize:

Doctor Name

Specialization

Consultation Fee

Call the parent constructor.

Create a class Patient inheriting from Doctor.

Constructor should initialize:

Patient Name

Age

Disease

Number of Consultation Days

Call the parent constructor.

Display:

Hospital Information

Doctor Information

Patient Information

Total Consultation Charges

GST (18%)

Final Bill

```python
class Hospital():
    def __init__(self , hosp_name , city):
        self.hosp_name = hosp_name
        self.city = city
        
    def display_hosp_info(self):
        print(f"Hospital Name: {self.hosp_name}")
        print(f"City: {self.city}")
        
class Doctor(Hospital):
    def __init__(self , doc_name , specialization , fee , hosp_name , city):
        self.doc_name = doc_name
        self.specialization = specialization
        self.fee = fee
        super().__init__(hosp_name , city)
        
    def display_doc_info(self):
        print(f"Doctor Name: {self.doc_name}")
        print(f"Specialization: {self.specialization}")
        print(f"Fee: {self.fee}")
        print(f"Hospital Name: {self.hosp_name}")
        print(f"City: {self.city}")
        
class Patient(Doctor):
    def __init__(self , name , age , disease , days, doc_name , specialization , fee , hosp_name , city):
        self.name = name
        self.age = age
        self.disease = disease
        self.days = days
        super().__init__(doc_name , specialization , fee , hosp_name , city)
        
    def display_patient_info(self):
        print(f"Patient Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Disease: {self.disease}")
        print(f"Days: {self.days}")
        self.display_doc_info()
        
    def calculate_bill(self):
        total_bill = self.fee * self.days
        print(f"Total Bill: {total_bill}")
        
pat = Patient("John" , 45 , "Heart Disease" , 5 , "Dr. Smith" , "Cardiologist" , 5000 , "Apollo" , "Delhi")
pat.display_patient_info()
```
