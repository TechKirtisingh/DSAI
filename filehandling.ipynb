```python
def Add(a,b):
    return a+b
def Subtract(a,b):
    return a-b
def Multiply(a,b):
    return a*b  
def Divide(a,b):
    return a/b
def calculatepower(a,b):
    return a**b

a=int(input("Enter first number: "))
b=int(input("Enter second number: "))

ch=int(input("Enter your choice: press 1 for '+' ,\n press 2 for '-' ,\n press 3 for '*' ,\n press 4 for '/' ,\n press 5 for '**' : "))

if ch==1:
    a = Add(a,b)
    print(a)
    file = open("gkt.txt", "w")
    file.write(str(a))
    file.close()
elif ch==2:
    s = Subtract(a,b)
    print(s)
    file = open("gkt.txt", "w")
    file.write(str(s))
    file.close()
elif ch==3:
    m = Multiply(a,b)
    print(m)
    file = open("gkt.txt", "w")
    file.write(str(m))
    file.close()
elif ch==4:
    d = Divide(a,b)
    print(d)
    file = open("gkt.txt", "w")
    file.write(str(d))
    file.close()
elif ch==5:
    p = calculatepower(a,b)
    print(p)
else:
    print("Invalid choice")
```
