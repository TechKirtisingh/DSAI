```python
# constructor::

class Student :
    course ="Data Science"
    
    def __init__(self , rn , nam , c , p , ds , daa):
        self.rollno = rn
        self.name = nam
        self.CLang = c
        self.datascience = ds
        self.python = p
        self.DAA = daa
        
    def displayStudentReport(self):
        tot = self.CLang + self.python + self.datascience+self.DAA
        per = tot/300*100
        grade = "\n"
        
        if per>=80:
            print("grade A") # display
            grade = "A" # store
        elif per>=60:
            print("grade B")
            grade = "B" 
        elif per>=40:
            print("grade C")
            grade = "C"
        else:
            print("grade D")
            grade = "D"
            
        print("\n" + "=" *100)
        print("Student Report Card")
        print("=" *100)
        print(f"Roll No.: {self.rollno}                   Name: {self.name}")
        print("="*100)
        print(f"CLanguage: {self.CLang}")
        print(f"Python: {self.python}")
        print(f"Data Science: {self.datascience}")
        print(f"DAA: {self.DAA}")
        print("="*100)
        print(f"Total: {tot}           Percentage: {per}%                Grade: {grade}")
        print("="*100)
        
s1 = Student(1 , "Kriti" , 90 , 80 , 70 , 60)
s2 = Student(2 , "Rohit" , 30 , 70 , 60 , 50)
s1.displayStudentReport()
s2.displayStudentReport()
```

```python
# function overloading: is not directly supported in python, but we can achieve it by using default arguments or variable-length arguments.

def add(a,b):
    s1 = a+b
    print(f"Sum of two numbers  {a} and {b} is: {s1}")

def add(a,b,c):
    s1 = a+b+c
    print(f"Sum of three numbers  {a}, {b} and {c} is: {s1}")

a ,b , c = map(int , input("Enter three numbers: ").split())
add(a,b,c)
add(a,b)
```

```python
# function overloadding : definning multiple functions with the same name but different parameters. In python, the last defined function will override the previous ones.

def addnumber(*args): # *args is a variable-length argument that allows us to pass any number of arguments to the function.
    return sum(args)

print(f"Sum of two numbers: {addnumber(10,20)}")
print(f"Sum of three numbers: {addnumber(10,20,30)}")
print(f"Sum of four numbers: {addnumber(10,20,30,40)}")
```

```python
# private

class Vechicle:
    def __init__(self , name , model):
        self.name = name  # public variable
        self._model = model # protected variable
        self.__speed = 0 # private variable

    def setSpeed(self , speed):
        self.__speed = speed

    def getSpeed(self):
        return self.__speed 
```
