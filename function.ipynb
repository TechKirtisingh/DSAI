```python
def Add(a,b):
    return a+b
def Subtract(a,b):
    return a-b
def Multiply(a,b):
    return a*b  
def Divide(a,b):
    return a/b
def calculatepower(a,b):
    return a**b

a=int(input("Enter first number: "))
b=int(input("Enter second number: "))

ch=int(input("Enter your choice: press 1 for '+' ,\n press 2 for '-' ,\n press 3 for '*' ,\n press 4 for '/' ,\n press 5 for '**' : "))

if ch==1:
    s = Add(a,b)
    print(s)
elif ch==2:
    s = Subtract(a,b)
    print(s)
elif ch==3:
    s = Multiply(a,b)
    print(s)
elif ch==4:
    s = Divide(a,b)
    print(s)
elif ch==5:
    s = calculatepower(a,b)
    print(s)
else:
    print("Invalid choice")
    

    
```

```python
def fact(num):
    if num==0 or num == 1:
        return 1
    else:
        return num*fact(num-1)
    
def min(n1 , n2):
    if n1>n2:
        return n2
    else:
        return n1
    
def Avg(n1 , n2):
    return (n1+n2)/2

def Even(num):
    if num%2==0:
        print("Even number")
    else:
        print("Odd number")

def Max(n1 , n2):
    if n1>n2:
        return n1
    else:
        return n2
    
num = int(input("Enter a number: "))
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))

ch = int(input("Enter your choice: press 1 for 'factorial' ,\n press 2 for 'minimum' ,\n press 3 for 'average' ,\n press 4 for 'even/odd' ,\n press 5 for 'maximum' : "))

if ch==1:
    s=fact(num)
    print(f"Factorial of {num} is: {s}")
    
elif ch==2:
    s=min(n1,n2)
    print(f"Minimum of {n1} and {n2} is: {s}")
    
elif ch==3:
    s=Avg(n1,n2)
    print(f"Average of {n1} and {n2} is: {s}")
    
elif ch==4:
    Even(num)
    print(f"Checking if {num} is even or odd:")

elif ch==5:
    s=Max(n1,n2)
    print(f"Maximum of {n1} and {n2} is: {s}")
    
else:
    print("Invalid choice")
```
