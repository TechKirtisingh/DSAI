```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Brand": ["Toyota", "Honda", "Hyundai", "Tata", "Mahindra", "Kia"],
    "Sales": [120, 95, 110, 85, 100, 90],
    "Price": [12, 10, 11, 9, 14, 13],
    "Mileage": [18, 20, 19, 22, 17, 18]
}

df = pd.DataFrame(data)

plt.figure(figsize=(14, 10))

#Graph 1: Sales by Brand
plt.subplot(2, 2, 1)
sns.barplot(data=df, x='Brand' , y='Sales', palette='pastel')
plt.title('Car Sales by Brand')
#Graph 2: Price vs Sales
plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='Price', y='Sales', hue='Brand', palette='pastel', s=100)
plt.title('Price vs Sales')
#Graph 3: Histogram of Mileage
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='Mileage', bins=6, kde=True, color='orange')
plt.title('Mileage Distribution')
#Graph 4: Box Plot of Sales
plt.subplot(2, 2, 4)
sns.boxplot(data=df, y='Sales', color='skyblue')
plt.title('Sales Outlier Detection')
```

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Patient": ["P1","P2","P3","P4","P5","P6","P7","P8"],
    "Age": [25,40,35,60,45,30,55,50],
    "BloodPressure": [120,140,130,160,150,125,155,145],
    "Sugar": [90,110,105,180,150,95,170,140]
}

df = pd.DataFrame(data)

plt.figure(figsize=(14, 10))

#Graph 1: Patient vs Blood Pressure
plt.subplot(2, 2, 1)
sns.lineplot(data=df, x='Patient', y='BloodPressure', marker='o', color='blue')
plt.title("Blood Pressure of Patients")

#Graph 2: Age vs Sugar Levels
plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='Age', y='Sugar', hue='Patient', palette='pastel', s=100)
plt.title("Age vs Sugar Levels")

#Graph 3: Histogram of Blood Pressure
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='BloodPressure',bins=5, kde=True, color='purple')
plt.title("Blood Pressure Distribution")

#Graph 4: Box Plot of Sugar Levels
plt.subplot(2, 2, 4)
sns.boxplot(data=df, y='Sugar', color='lightgreen')
plt.title("Sugar Level Outlier Detection")
```
