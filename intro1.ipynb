```python
# types of data type
# 1. Integer    2. Float    3. String    4. Boolean 

# way to declare String
# 1. Single quote       2. Double quote       3. Triple quote

s1= 'RAM SETU'
s2= "GYAN SETU"    
s3= """RAM SETU IS A BRIDGE BUILT BETWEEN INDIA AND SRI LANKA"""
s4= '''GYAN SETU IS A INSTITUTE'''     


print(s1 , type(s1)  , s2 , type(s2)  , s3 , type(s3)  , s4 , type(s4))
```

```python
# slicing of string
# star index include and end index exclude 
# :third parameter is used to denote how many characters to skip while slicing the string (by default it is 1)


s= "RAM SETU"


print(s[0:3])  
print(s[-1])   # finding the last character of string
print(s[0:3:1]) # 0 to 3 index with step size 1
print(s[2:6:2]) # it pr
print(s[-1:-5:-1]) # reverse slicing of string
print(s[::-1]) # reverse the string
print(s[-8: -2])
```

```python
# LIST

mylist =[12 , 34.5 , "python" , True]

print(mylist)

print(type(mylist))

print(mylist[-3])

print(mylist[1:3:1])
```

```python
# nested list

mylistn =[[1,2,3] , ["a" , "b" , "c"] , [1.1 , 2.2 , 3.3]]

print(mylistn)

print(mylistn[1][2])  # accessing the nested list

print(mylistn[1][2])

print(mylistn[2][2])  # it will give the last element of the nested list
```
